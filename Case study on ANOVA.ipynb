3).To check whether all the zones generate the same amount of sales.

H0:All the zones generate the same amount of sales
H1:All the zones generate different amount of sales

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
fvalue,pvalue=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print(fvalue,pvalue)
```

```python
*Here the P-value is 0.00118 which is lower than the significance level 0.05 so we reject the null hypothesis and 
 we get that all the zones generate different amount of sales. 
```
